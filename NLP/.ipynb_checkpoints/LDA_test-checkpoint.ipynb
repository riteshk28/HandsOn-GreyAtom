```python
import pandas as pd
import numpy as np
import nltk
from nltk.tokenize import RegexpTokenizer, treebank
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
tokenizer = RegexpTokenizer(r'\w+')
en_stop = set(stopwords.words('english'))
p_stemmer = PorterStemmer()
```

```python
with open("./LDA_test.txt", "r") as f:
    f1 = f.read()
```

```python
texts = []
```

```python
for i in sents:
    
    # clean and tokenize document string
    raw = i.lower()
    tokens = tokenizer.tokenize(raw)
    # remove stop words from tokens
    stopped_tokens = [i for i in tokens if not i in en_stop]
    # stem tokens
    stemmed_tokens = [p_stemmer.stem(i) for i in stopped_tokens]
    # add tokens to list
    texts.append(stemmed_tokens)
```

```python
texts[:2]
```

```python
NUM_TOPICS = 3
topic_allocation = []
```

```python
for text in texts:
    topic_allocation.append([(i, np.random.randint(NUM_TOPICS)) for i in text])
```

```python
topic_allocation
```

```python
token_list = []
topic_list = []
doc_list = []
for i, text in enumerate(topic_allocation):
    for token in text:
        topic_list.append(token[1])
        token_list.append(token[0])
        doc_list.append(i)
```

```python
df = pd.DataFrame({"token": token_list,
             "topic_no": topic_list,
             "doc_no": doc_list})
```

```python
df.head(20)
```

```python
token_dist = df.pivot_table(values=["topic_no"], index=["token"], columns=["topic_no"], aggfunc="count")
token_dist.reset_index(inplace=True)
token_dist.rename(columns={0: "grp_1", 1: "grp_2", 2: "grp_3"}, inplace=True)
token_dist.fillna(0, inplace=True)
token_dist.head()
```

```python
doc_dist = df.pivot_table(values=["topic_no"], index=["doc_no"], columns=["topic_no"], aggfunc="count")
doc_dist.reset_index(drop=True, inplace=True)
doc_dist.rename(columns={0: "grp_1", 1: "grp_2", 2: "grp_3"}, inplace=True)
doc_dist.fillna(0, inplace=True)
doc_dist.head()
```

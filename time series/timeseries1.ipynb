```python
import numpy as np
import pandas as pd
import datetime

tss = pd.read_csv("./data/AusAntidiabeticDrug.csv")
tss["ds"] = pd.to_datetime(tss["ds"])
tss.index = tss["ds"]
tss = tss[["y"]]
tss.head()
tss["Trend"] = [i for i in range(len(tss))]
## Split the data into train and test
x_train = tss[tss.index <datetime.datetime(2005, 1, 1)]
x_test = tss[tss.index >= datetime.datetime(2005, 1, 1)]
print(x_train.shape, x_test.shape)
```

```python
class feature_engineer:
    
    def __init__(self, X_train, X_test):
        self.X_train = X_train.Trend.reshape(-1, 1)
        self.y_train = X_train.y
        self.X_test = X_test.Trend.reshape(-1, 1)
        self.y_test = X_test.y
#         self.X_test=X_test
```

```python
class timeseries_lr:
    
    def __init__(self, model):
        self.test = "test"
        self.model = model
        self.predictions = None
    
    def fit(self, x, y):
        self.model.fit(x, y)
    
    def predict(self,x_test):
        self.predictions = self.model.predict(x_test)
        
```

```python
fe = feature_engineer(x_train, x_test)
```

```python
from sklearn.linear_model import LinearRegression
model = timeseries_lr(LinearRegression())
model.fit(fe.X_train, fe.y_train)
model.predict(fe.X_test)
```
